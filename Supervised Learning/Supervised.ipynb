## PreProcessing

```python
import numpy as np
from scipy import stats
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')

filtered_entries = np.array([True] * len(df))

for col in ['comment_count', 'dislikes', 'likes']:
    zscore = abs(stats.zscore(df[col])) # hitung absolute z-scorenya
    filtered_entries = (zscore < 3) & filtered_entries # keep yang kurang dari 3 absolute z-scorenya
    
df_prep= df[filtered_entries] # filter, cuma ambil yang z-scorenya dibawah 

print(f'Jumlah baris setelah memfilter outlier: {len(df_prep)}')
```

```python
df_prep
df_prep = df_prep.drop(columns=categorical)
df_prep
```

```python
df_prep= df_prep.drop(columns='publish_date')
df_prep
```

# Hyperparameter Tuning decision Tree regressor (prevent overfitting)
according to https://towardsdatascience.com/how-to-tune-a-decision-tree-f03721801680 ways to reduce overfitting on dt model:
* set `splitter` parameter as `random`
* reducing number of `max_depth`
* `max features` as log2(n_features)

```python
def hyperdt_evaluation(X,y):
    x_train ,x_test,y_train,y_test = train_test_split(X,y,test_size=1/3, random_state=42)
    max_depth = [int(x) for x in np.linspace(1, 110, num = 30)] # Maximum number of levels in tree
    min_samples_split = [2, 5, 10, 100] # Minimum number of samples required to split a node
    min_samples_leaf = [1, 2, 4, 10, 20, 50] # Minimum number of samples required at each leaf node
    max_features = ['auto', 'sqrt'] # Number of features to consider at every split
    splitter = ['random']
    hyperparameters = dict(max_depth=max_depth, 
                       min_samples_split=min_samples_split, 
                       min_samples_leaf=min_samples_leaf,
                       max_features=max_features,
                       splitter=splitter
                      )
    
    dt = DecisionTreeRegressor(random_state=42)
    model = RandomizedSearchCV(dt, hyperparameters, cv=5, random_state=42)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    pred_train = model.predict(x_train)
    print('R2 score test: %.2f' % r2_score(y_test, pred))
    print('R2 score train: %.2f' % r2_score(y_train, pred_train))
```

```python
for x in x_values:
    print(hyperdt_evaluation(x,y))
```
